## Data loading

### Subtask:
Load the Olympic medalists dataset.


**Reasoning**:
Load the dataset and display the first few rows along with its shape to verify successful loading.



```python
import pandas as pd

try:
    df_medalists = pd.read_csv('Olympic_Medalists_Cleaned (1).csv')
    display(df_medalists.head())
    print(df_medalists.shape)
except FileNotFoundError:
    print("Error: 'Olympic_Medalists_Cleaned (1).csv' not found. Please ensure the file exists in the current directory or provide the correct path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data cleaning

### Subtask:
Clean the Olympic medalists dataset.


**Reasoning**:
Inspect the DataFrame for missing values and anomalies, then clean the data accordingly. This involves checking for missing values, handling them appropriately, and addressing inconsistencies in the data.



```python
# Check for missing values
missing_values = df_medalists.isnull().sum()
print("Missing values per column:\n", missing_values)

# Check for duplicates
duplicates = df_medalists.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Handle missing values (example: fill missing 'Medal' with 'No Medal')
df_medalists['Medal'].fillna('No Medal', inplace=True)

#Standardize format of certain columns
df_medalists['Gender'] = df_medalists['Gender'].str.capitalize()
df_medalists['Event_gender'] = df_medalists['Event_gender'].str.upper()

#Further inspection for anomalies and inconsistencies
#Example: Check for inconsistent capitalization in 'Sport' and 'Discipline'
print("\nUnique values of 'Sport':\n", df_medalists['Sport'].unique())
print("\nUnique values of 'Discipline':\n", df_medalists['Discipline'].unique())

#Further cleaning steps can be added here based on the inspection results.

# Display the first few rows and shape after cleaning
display(df_medalists.head())
print(df_medalists.shape)
```

## Data wrangling

### Subtask:
Data Wrangling: Convert categorical variables and create necessary calculated fields.


**Reasoning**:
Convert specified columns to categorical type, convert 'Edition' to numeric type, create a 'Decade' column, and verify the data types using df_medalists.info().



```python
# Convert specified columns to categorical
categorical_cols = ['Medal', 'Gender', 'Sport', 'NOC', 'Event_gender']
for col in categorical_cols:
    df_medalists[col] = pd.Categorical(df_medalists[col])

# Convert 'Edition' to numeric
df_medalists['Edition'] = pd.to_numeric(df_medalists['Edition'], errors='coerce')

# Create 'Decade' column
df_medalists['Decade'] = (df_medalists['Edition'] // 10) * 10

# Verify data types
df_medalists.info()
```

## Data preparation

### Subtask:

## Data exploration

### Subtask:
Explore the prepared data to verify data cleaning and wrangling.


**Reasoning**:
Explore the prepared data to verify data cleaning and wrangling by checking the shape, data types, value distributions, and descriptive statistics.



```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", df_medalists.shape)

# Verify data types
print("\nData Types of Columns:\n", df_medalists.info())

# Examine the distribution of key variables
print("\nValue Counts for 'Medal':\n", df_medalists['Medal'].value_counts())
print("\nValue Counts for 'Gender':\n", df_medalists['Gender'].value_counts())
print("\nValue Counts for 'Sport':\n", df_medalists['Sport'].value_counts())
print("\nValue Counts for 'NOC':\n", df_medalists['NOC'].value_counts())

# Calculate descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:\n", df_medalists.describe())

# Check unique values for 'Decade'
print("\nUnique values in 'Decade' column:\n", df_medalists['Decade'].unique())

#Check the min and max values of the Edition column
print("\nMin and Max values of Edition column:\n", df_medalists['Edition'].min(), df_medalists['Edition'].max())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Edition and Gender, then count the number of athletes
gender_evolution = df_medalists.groupby(['Edition', 'Gender'])['Athlete'].count().reset_index()

# Pivot the table to have separate columns for 'Men' and 'Women' counts
gender_evolution_pivot = gender_evolution.pivot(index='Edition', columns='Gender', values='Athlete').fillna(0)

# Plotting the dual-line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=gender_evolution_pivot, markers=True)
plt.title('Evolution of Athlete Participation by Gender Over Time')
plt.xlabel('Edition')
plt.ylabel('Number of Athletes')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Sport and Medal, then count the number of medals
sport_medal_counts = df_medalists.groupby(['Sport', 'Medal']).size().reset_index(name='Medal_Count')

# Filter out 'No Medal' if it was used as a fill value during cleaning
sport_medal_counts = sport_medal_counts[sport_medal_counts['Medal'] != 'No Medal']

# Pivot the data to get Medal types as columns
sport_medal_pivot = sport_medal_counts.pivot(index='Sport', columns='Medal', values='Medal_Count').fillna(0)

# Reorder columns for consistent stacking (e.g., Bronze, Silver, Gold)
medal_order = ['Bronze', 'Silver', 'Gold']
sport_medal_pivot = sport_medal_pivot[medal_order]

# Sort sports by total medal count
sport_medal_pivot['Total'] = sport_medal_pivot.sum(axis=1)
sport_medal_pivot = sport_medal_pivot.sort_values('Total', ascending=False).drop('Total', axis=1)

# Plotting the stacked bar chart
plt.figure(figsize=(14, 8))
sport_medal_pivot.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Medal Distribution by Sport and Medal Type')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.xticks(rotation=90)
plt.legend(title='Medal Type')
plt.tight_layout()
plt.show()
```
